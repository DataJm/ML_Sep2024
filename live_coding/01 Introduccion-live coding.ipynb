```python
import pandas as pd
```

```python
data = pd.read_csv("../data/german.csv")
data
```

```python
# Vamos a crear un modelo de regresión lineal. Una regresión lineal es:
# Modelo supervizado, modelo de regresión

# Todo lo que yo crea que me sirve para predecir la variable "y" va en Equis (X)
X = data[['Age']] # X (mayúscula) es el CONJUNTO de todas mis variables predictoras

# La variable "y" (minúscula) es la variable que queremos predecir. 
y = data['CreditAmount'] # La variable y es la variable objetivo (por eso es una columna -serie)
```

```python
from sklearn.linear_model import LinearRegression
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
# Los modelos de sklearn, son objetos (tienen sus métodos y sus propiedades)
modelo = LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
# Ahora que nuestro modelo ha aprendido la cantidad de crédito que se le da a las personas
# en función de su edad, pues vamos a usarlo para hacer nuevas predicciones.

# Para esto, necesitamos hablarle al modelo en el único idioma que entiende:
# un dataframe con la columna Age
simulacion = pd.DataFrame({
    'Age': [20,30,40,50]
})

simulacion
```

```python
modelo.predict(simulacion)
```

# Cuál es la matemática detrás del modelo y cómo es diferente ML de la estadística "tradicional" 

A continuación tenemos un código más "tradicional" de hacer regresión lineal. 

```python
import statsmodels.api as sm
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

```python
modelo.score(X,y)
```
